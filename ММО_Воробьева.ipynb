```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('diabetes.csv')

sns.pairplot(data, kind="scatter", hue="Outcome", plot_kws=dict(s=100, edgecolor="white", linewidth=2))

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("diabetes.csv")

fig = plt.figure(figsize=(10,10), dpi=50)
graph = fig.add_subplot()

class_counts = data['Outcome'].value_counts()
bars = graph.bar(class_counts.index, class_counts.values, color=["yellow", 'green'])

plt.gca().set(
    xticks=range(3),
    xlabel='Class',
    ylabel='Count'
)

plt.title("Classes allocation")
plt.legend(bars, ["Sick", "Not sick"])

plt.show()
```
